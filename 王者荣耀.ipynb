# 处理数据（一）

## 数据清洗

### 1.检测缺失值

```python
df.isnull()
```

### 2.计算缺失值数量

```python
df.isnull().sum()
```

### 3.填充缺失值

```python
df.fillna(
    value={                         #要填充的数据
        'critical_hit': 0,             
        'attack_speed': 0,
        'magic_penetration': 0,
        'physical_penetration': 0,
        'magic_attack': 0
    },
    inplace=True                   #原数据补全
)
```

```python
df.isnull().sum()
```

### 4.把旧内容替换成新内容

```python
df['hero_attr_difficulty'] = df['hero_attr_difficulty'].str.replace('star-', '')
df['hero_attr_difficulty']
```

### 5.按列排序(单列排序)

```python
df.sort_values(by='hero_attr_difficulty', inplace=True)
df['hero_attr_difficulty']
```

### 6.把XX转换为XX类型

```python
df['hero_attr_difficulty'] = df['hero_attr_difficulty'].astype('int')     
df['hero_attr_difficulty']
```

```python
df.sort_values(by='hero_attr_difficulty', inplace=True)
df['hero_attr_difficulty']
```

```python
df.head(10)
```

# 处理数据（二）

### 1.加权平均数

```python
df['ability'] = df['max_life'] * 0.2 + df['physical_attack'] * 0.3 + df['magic_attack'] * 0.2 + df['physical_defense'] * 0.15 + df['magic_defense'] * 0.15
df
```

```python
df.sort_values(by='ability',inplace=True, ascending=False)
df
```

```python
df.head(10)
```

### 2.按列排序(多列排序)

```python
df.sort_values(
    by=['max_life',
        'physical_attack',
        'magic_attack',
        'physical_defense',
        'magic_defense'
       ],
    inplace=True,
    ascending=False
)
df
```

```python
df.head(10)
```

### 3.修改错误值

```python
df.loc[25, 'physical_attack'] = 166
df.loc[25, 'physical_attack']
```

```python
df.head(10)
```

### 2.绘制饼状图

```python
x =  df['hero_type'].value_counts()        #各英雄类型个数的统计值
labels = set(df['hero_type'])              #英雄类型名称
autopct = '%2.1f%%'                        #百分数显示格式,保留1位小数
plt.figure(figsize=(8, 6), dpi=100)
patches, l_text, p_text = plt.pie(x=x, labels= labels, autopct=autopct)
for t in l_text:                           #遍历labels,设置为font_set
    t.set_fontproperties(font_set)
plt.title('各英雄类型占比', fontproperties=font_set)
plt.show()
```

### 3.绘制折线图

# 数据可视化（二）

### 1.绘制多条形图

```python
data1 = df.loc[80,['physical_attack', 'magic_attack', 'physical_defense', 'magic_defense']]
data2 = df.loc[99,['physical_attack', 'magic_attack', 'physical_defense', 'magic_defense']]
plt.figure(
figsize=(10, 8), 
dpi=100
)
plt.bar(data1.index, data1.values)
plt.bar(data2.index, data2.values)
plt.show()
```

```python
bar_width = 0.25  # 设置条形图宽度
x_1 = []  # data1对应的x坐标
x_2 = []  # data2对应的x坐标
for i in range(len(data1)):
    x_1.append(i+bar_width)
    x_2.append(i+bar_width * 2)
plt.figure(
figsize=(10, 8), 
dpi=100
) 
plt.bar(x_1, data1.values, width=bar_width, label='廉颇')#设置图例
plt.bar(x_2, data2.values, width=bar_width, label='亚瑟') #设置图例
u = []
for i in range(len(x_1)):
    x = (x_1[i] + x_2[i]) / 2
    u.append(x)
plt.xticks(u, data1.index, size=16)
plt.title('属性对比', fontproperties=font_set)
plt.legend(
    prop=font_set
)
plt.show()
```

### 2.绘制多折线图

```python
label_list = df.head(10)['hero_name']  # 横坐标刻度显示值
y1 = df.head(10)['physical_attack']
y2 = df.head(10)['magic_attack']
y3 = df.head(10)['physical_defense']
y4 = df.head(10)['magic_defense']

plt.plot(label_list, y1, color='blue', linewidth=1.0, linestyle='--')
plt.plot(label_list, y2, color='blue', linewidth=1.0, linestyle='-.')
plt.plot(label_list, y3, color='green', linewidth=1.0, linestyle='--')
plt.plot(label_list, y4, color='green', linewidth=1.0, linestyle='-.')

plt.title("能抗能打TOP10属性对比", fontproperties=font_set)
plt.legend(labels=['physical_attack', 'magic_attack', 'physical_defense', 'magic_defense'])
plt.xticks(fontproperties=font_set)
plt.show()
```

### 3.绘制复合图表

```python
label_list = df.head(10)['hero_name']  # 横坐标刻度显示值
y1 = df.head(10)['physical_attack']
y3 = df.head(10)['physical_defense']

plt.bar(label_list, y1, color='#EEE',label='physical_attack')
plt.plot(label_list,y3,color='green',linewidth=1.0,linestyle='--', label='physical_defense')

plt.title("能抗能打TOP10属性对比", fontproperties=font_set)
plt.legend()
plt.xticks(fontproperties=font_set)
plt.show()
```
